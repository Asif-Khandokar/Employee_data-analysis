Load the TrainingHours column into a NumPy array.
         Convert to float datatype
         Find mean & standard deviation
         Final output: Two numeric values

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/final-employee-ds.csv')
```

```python
df
```

```python
training_np = df["TrainingHours"].to_numpy()
training_np = training_np.astype(float)
mean_value = training_np.mean()
std_value  = training_np.std()

print(mean_value)
print(std_value)
```

```python
print(df["Salary"].isna().sum())
```

```python
dept_median = df.groupby("Department")["Salary"].median()
df["Salary"] = df["Salary"].fillna(df["Department"].map(dept_median))
print(df["Salary"].isna().sum())
```

```python
filtered = df[(df["ExperienceYears"] > 12) & (df["ProjectCount"] > 5)]
total_salary_sum = filtered["Salary"].sum()
print(total_salary_sum)
```

Answer to the Question No 3

```python
dept_filter = df["Department"].isin(["IT", "Finance"])
age_filter = df["Age"].between(30, 45)
perf_filter = df["PerformanceScore"] > 88
filtered_df = df[dept_filter & age_filter & perf_filter]
filtered_df = filtered_df.sort_values("Salary", ascending=False)
```

```python
new_df = filtered_df[["Name", "Department", "Salary", "PerformanceScore"]]
print(new_df)
```

Answer to the Question No 5

```python
counts = (
    df.groupby("Department")["Gender"]
      .value_counts()
      .reset_index(name="Count")
)


male = counts[counts["Gender"] == "Male"].set_index("Department")["Count"]
female = counts[counts["Gender"] == "Female"].set_index("Department")["Count"]

departments = sorted(df["Department"].unique())
male = male.reindex(departments, fill_value=0)
female = female.reindex(departments, fill_value=0)

plt.bar(departments, male, label="Male")
plt.bar(departments, female, bottom=male, label="Female")

plt.title("Employee Count by Department (Split by Gender)")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.legend(title="Gender")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
